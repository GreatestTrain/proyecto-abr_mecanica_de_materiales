```python
from matplotlib.pylab import *
import matplotlib as mpl
```

```python
def f(x, y):
    return x**2 + y**2
```

```python
res = 10
x = linspace(-5,5,res)
y = linspace(-5,5,res)
x, y = meshgrid(x,y)
z = f(x,y)
```

```python
fig, ax = subplots()
level_map = np.arange(np.min(z), np.max(z), 5)
color_map = ax.contour(x,y,z, levels=level_map, cmap = cm.inferno)
fig.colorbar(color_map)
show()
```

```python
fig, ax = plt.subplots(subplot_kw={'projection':'3d'}, figsize=(6,6))
surf = ax.plot_surface(x,y,z, cmap=cm.inferno)
fig.colorbar(surf)
plt.show()
```
